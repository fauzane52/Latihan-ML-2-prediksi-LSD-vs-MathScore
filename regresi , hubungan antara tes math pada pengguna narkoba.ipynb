```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv('/content/lsd_math_score_data.csv')
data
```

```
#ada 7 orang di untk narkoba
#times delay in minutes, berarti data menit setelah disuntikkan narkoba
#avg math score berarti nilainya
#LSD ppm berarti jumlah part per million nya narkoba yang di suntik

#data dibawah merupakan series, ganti ke Dataframe agar bisa di masukan ke regressi
waktu = data[['Time_Delay_in_Minutes']]
narkoba=data[['LSD_ppm']]
nilai=data[['Avg_Math_Test_Score']]
```

```
#visualisasi data
plt.plot( waktu , narkoba , 'r', linewidth=3)

#kasih judul
plt.title('narkoba yang disuntikin ke tubuh vs waktu', fontsize=15)

#kasih label x dan y
plt.xlabel('waktu dalam menit',fontsize=10)
plt.ylabel('LSD disuntikkan PPM', fontsize=10)

#kasih tebel angka di x dan y
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

#kasih limit
plt.ylim(1,7)
plt.xlim(0,500)


plt.style.use('classic')

#kasih footer
plt.text(x=0,y=-0.2,s='Wagner et al (1968)')
plt.show()
```

```
from sklearn.linear_model import LinearRegression
regr=LinearRegression()

#memasukan ke linear regression dengan series data, jadi ubah dlu ke dataframe
regr.fit( narkoba,nilai)

print ('Theta1 :',regr.coef_ [0][0])  #mengambil nilai coef_ atau m atau theta 1

print ('Theta0 or intercept :',regr.intercept_[0])    #mengambil koefisien C atau Theta 0 dari persamaan y=mx + C

#score validasi

print('R-square adalah',regr.score(narkoba,nilai))

#prediksi regresi

prediksi_nilai=regr.predict(narkoba)
```
